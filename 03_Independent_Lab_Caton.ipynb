## Please Enjoy our Vending Machine and Have a Nice Day!

```python
print("Welcome to the snack and drink dispenser for desperate and tired parents.  We hope you enjoy our selection of kid snacks and beverages")

skit = "Skittles"
chee = "Cheetos"
slic = "Apples"
coff = "Coffee"
rbul = "Red Bull"
skitprice = 2.32
cheeprice = 3.72
slicprice = 1.55
coffprice = 2.09
rbulprice = 4.99
skitnum = "1."
cheenum = "2."
slicnum = "3."
coffnum = "4."
rbulnum = "5."
skitstr = str(skitprice)
cheestr = str(cheeprice)
slicstr = str(slicprice)
coffstr = str(coffprice)
rbulstr = str(rbulprice)



totalCost = 0.0
totalskit = 0
totalchee = 0
totalslic = 0
totalcoff = 0
totalrbul = 0

i=0
while i== 0:
    print(f"{skitnum} {skit}: ${skitstr}")
    print(f"{cheenum} {chee}: ${cheestr}")
    print(f"{slicnum} {slic}: ${slicstr}")
    print(f"{coffnum} {coff}: ${coffstr}")
    print(f"{rbulnum} {rbul}: ${rbulstr}")
    selection = int(input("Please select the item number you wish to purchase\n"))
    quantity = 0
    quantskit = 0
    quantchee = 0
    quantslic = 0
    quantcoff = 0
    quantrbul = 0  
    quantity = int(input("Please input the quantity you wish to purchase\n"))
    if selection == 1:
        quantskit = quantskit + quantity
        totalskit = totalskit + quantity
    elif selection == 2:
        quantchee = quantchee + quantity
        totalchee = totalchee + quantity
    elif selection == 3:
        quantslic = quantslic + quantity
        totalslic = totalslic + quantity
    elif selection == 4:
        quantcoff = quantcoff + quantity
        totalcoff = totalcoff + quantity
    elif selection == 5:
        quantrbul = quantrbul + quantity
        totalrbul = totalrbul + quantity
    else:
        print("That is not an Option, Please Try Again")
    totalCost = totalCost + (quantskit*skitprice)+(quantchee*cheeprice)+(quantslic*slicprice)+(quantcoff*coffprice)+(quantrbul*rbulprice)
    print(f"{skitnum} {skit}: ${skitstr}")
    print(f"{cheenum} {chee}: ${cheestr}")
    print(f"{slicnum} {slic}: ${slicstr}")
    print(f"{coffnum} {coff}: ${coffstr}")
    print(f"{rbulnum} {rbul}: ${rbulstr}")
    print("Total Cost $",round(totalCost, 2))
    checkout = int(input("Would you like to Checkout? 0 for No, 1 for Yes, Other Input Cancels Order\n"))
    i = i + checkout
print("Please insert $",round(totalCost,2),"and the following items will dispense:")
print(totalskit,skit)
print(totalchee,chee)
print(totalslic,slic)
print(totalcoff,coff)
print(totalrbul,rbul)
```
